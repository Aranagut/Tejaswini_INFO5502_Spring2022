***Dropping Duplicates***

In the dataset we have city and location, we will keep one and drop another.

```python
df = df.drop(['listed_in(city)'], axis = 1)
```

```python
df.head()
```

***Cleaning Location Column***

```python
location = df['location'].value_counts(ascending  = False)

location_lessthan500 = location[location<500]



def handle_location(value):
    if(value in location_lessthan500):
        return 'others'
    else:
        return value
        
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

Cleaning Cuisines Column

```python
cuisines = df['cuisines'].value_counts(ascending  = False)


cuisines_lessthan100 = cuisines[cuisines<100]



def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

We have collected all the articles from AMIA in the last 30 years. The dataset is avaliable on GitHub: https://raw.githubusercontent.com/unt-iialab/info5502-spring2022/main/datasets/lab_assignment_05_amia_cleaned.csv.

```python
# Your python code here:

import pandas as pd
import matplotlib.pyplot as plt
import yake

df_file1 = pd.read_csv("https://raw.githubusercontent.com/unt-iialab/info5502-spring2022/main/datasets/lab_assignment_05_amia_cleaned.csv", sep=",")

df_file1 = df_file1[df_file1['year'].isna()!=True]

df_file1['year'] = df_file1['year'].astype('int64')

df_file1.groupby(df_file1['year']).head()
```

```python
count = 2
for year in df_file1['year'].unique():
    print(year)

    max_ngram_size = 4

    lang = "en"

    numOfKeywords = 6

    text = " ".join(title for title in df_file1[df_file1['year']==year]['title'].values)
    custom_kw_extractor = yake.KeywordExtractor(lan=lang, top=numOfKeywords, features=None)
    keywords = custom_kw_extractor.extract_keywords(text)

    for kyw in keywords:
        print(kyw)
    print("------------------")
    count += 2
```
